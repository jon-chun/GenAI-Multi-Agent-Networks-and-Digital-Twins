## Chunking data and extracting it with metadata

## Detailed word-by-word extraction

```python
md_text_words = pymupdf4llm.to_markdown(
    doc=filename_pdf,
    pages=[0, 1, 20, 21, 22],
    page_chunks=True,
    write_images=True,
    image_path="images",
    image_format="png",
    dpi=300,
    extract_words=True
)
```

```python
type(md_text_words)
```

```python
type(md_text_words[4])
```

```python
print(json.dumps(md_text_words[3], indent=2))
```
