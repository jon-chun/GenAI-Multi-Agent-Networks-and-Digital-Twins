## High Level Info
- Multimodal model can process text and image simulatenously

- Can process images of 1024*1024

- Context lengh of 128k tokens

- Suitable for OCR and info extraction

```python
%%time

# NOTE: 1m 58s

llm = LLM(
    model = "mistralai/Pixtral-12B-2409",
    tokenizer_mode= "mistral",
    max_model_len = 4000
)
```

```python
from vllm.sampling_params import SamplingParams
```

```python
prompt = "Can you explain the image and list down all the findings."
```

```python
import requests
```

```python
!pip install Pillow
```

```python
# 3. Convert the downloaded PNG file to a JPG file
from PIL import Image
```

# Image #1

```python
image_url =  "https://raw.githubusercontent.com/AIAnytime/Image-Repo/main/nvidia.png"
```

```python
image_url
```

```python
print("Original PNG image:")
display(IPImage(image_url))
```

```python
messages = [
    {
        "role": "user",
        "content": [
            {"type": "text", "text": prompt},
            {"type": "image_url", "image_url": {"url": image_url}}
        ]
    }
]
```

```python
messages
```

```python
%%time

# NOTE: 13s

outputs = llm.chat(
    messages,
    sampling_params=SamplingParams(max_tokens=8192)
)
```

```python
print(outputs[0].outputs[0].text)
```
