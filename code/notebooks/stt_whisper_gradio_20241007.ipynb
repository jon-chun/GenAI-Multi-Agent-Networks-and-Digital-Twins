```python
import torch
from transformers import pipeline
import gradio as gr
```

```python
pipe = pipeline("automatic-speech-recognition",
               "openai/whisper-large-v3-turbo",
               torch_dtype=torch.float16,
               device="cuda:0")

```

```python
# prompt: upload sound file with colab function and save upload file in upload_filename variable

from google.colab import files

uploaded = files.upload()
upload_filename = list(uploaded.keys())[0]
```

```python
pipe(upload_filename)
```

```python
def transcribe(inputs):
    if inputs is None:
        raise gr.Error("No audio file")

    text = pipe(inputs, generate_kwargs={"task": "transcribe"}, return_timestamps=True)["text"]
    return text
```

```python
demo = gr.Interface(
    fn=transcribe,
    inputs=[
        gr.Audio(sources=["microphone", "upload"], type="filepath"),
    ],
    outputs="text",
    title="Whisper Large V3 Turbo: Transcribe Audio",
    description=(
        "Transcribe long-form microphone or audio inputs. Thanks to HuggingFace"
    ),
    allow_flagging="never",
)

demo.launch()
```
